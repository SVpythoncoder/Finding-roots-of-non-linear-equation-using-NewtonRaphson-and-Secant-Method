```python
# IMPORTS
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from scipy import linalg

# MAIN
def LeastSqr(x,y,n):
    ym = np.mean(y)
    xmat,ymat = matrix(x,y,n)
    xmat = np.reshape(xmat,((n+1),(n+1)))
    ymat = np.reshape(ymat,((n+1),1))
    coeff = np.linalg.solve(xmat,ymat)
    Rsq = findRq(y,ym,np.array(fn(x,coeff)))
    return n,coeff,Rsq

# SOlVING
def matrix(x,y,n):
    xmat = np.array([]) 
    ymat = np.array([])
    for i in range(0,n+1):
        for j in range(0,n+1):
            v1 = np.sum(x**(i+j))
            xmat = np.append(xmat,v1)
        v2 = np.sum(y*(x**i))
        ymat = np.append(ymat,v2)
    return xmat,ymat
    
# FINDIND APPROX FUNCITON    
def fn(x, a):
    px = 0
    for index in range(0, np.size(a)):
        px += (a[index] * (x**index))   # evaluate the P(x)
    return px
    
# FINDING R-sq
def findRq(y,ym,fn):
    So = np.sum((y-ym)**2)
    S1 = np.sum((y-fn)**2)
    Rsq = (1-(S1/So))
    return Rsq

# FUNCITON FOR PLOTTING THE DATA
def plot(x,y,coeff):
    fig=plt.figure(figsize=(8, 8))
    fig.patch.set_facecolor('lightgray')
    fig.suptitle('LEAST SQUARE FITTING', fontsize=14, fontweight='bold')
    plt.axis([0.0,1.0,-5,25])
    plt.scatter(x, y, s=70, facecolors='yellow', edgecolors='red', label="Data")
    z=np.arange(0,1.01,0.01)
    plt.plot(z,fn(z,coeff), label="Fitted polynomial",color='green')
    plt.grid(True,color='black')
    plt.legend()
    plt.ylabel("Y")
    plt.xlabel('X')
    plt.show()


#INPUT
x,y = np.loadtxt('Testfile1.txt', # name of file
    skiprows = 1, # skip first row
    unpack = True # make each column an array
)                 
n =  int(input('Type the degree of the Approximate polynomial = '))
n,coeff,Rsq= LeastSqr(x,y,n)


# OUTPUT
f=open('OutputFile.txt', 'a')
f.truncate(0)  #to clear the file
f.write('Degree of the polynomial = %s\n\n'%n)
print('\n')
if n==1:
    f.write('Linear: ')
    print('Linear: ')
elif n==2:
    f.write('Quadratic: ')
    print('Quadratic: ')
elif n==3:
    f.write('Cubic: ')
    print('Cubic: ')
elif n==4:
    f.write('Quartic: ')
    print('Quartic: ')

f.write('Coefficients \n')
print('Coefficients \n')
coeff1 = np.zeros((n,1))
coeff1=coeff

# To remove brackets in output
for a in coeff1.flatten():
    f.write('%s\n' %round(a,3))
    print(round(a,3))
f.write('\nR-sq = %s'%round(Rsq,4))
f.close()
print('\nRsq = ',round(Rsq,4))

# PLOT
plot(x,y,coeff)
```
