```python
# PYHTON CODE FOR FINDING ROOTS USING NETWON RAPHSON AND SECANT METHOD
# CREATED BY SWAPIL VERMA 

# IMPORTS
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
import sympy as sy
from sympy.abc import _clash1

# MAIN BODY

def func( a,x ):
    d=a.subs('x',x)
    return d

def derivFunc( b,x ):
    f=b.subs('x',x)
    return f

def newtonRaphson( x,a,b,c,e ): 
    fa=[]
    fd=[]
    i=1
    E=0
    print('I.No '+'|\t'+'X'+'\t\t|\t\t'+'F(x)''\t\t|\t\t'+'F\'(x)'+'\t\t|\t'+'Ea'+'\t\t|')
    while i<=c:
        
        if i==1:
            print(''+str(i)+'    |\t'+str('%6f'%x)+'\t|\t'+str(func(a,x))+'\t|\t'+str(derivFunc(b,x))+'\t|\t'+str('%6f'%(E))+'\t|')
            
            # Plot between Func(x) vs X
            ax.scatter(x,func(a,x))
            
            # set a title and labels
            ax.set_title('F(x) vs x')
            ax.set_xlabel('x')
            ax.set_ylabel('F(x)')
            fa.append(i)
            fd.append(E)
            i+=1
        else:
            
            h = func(a,x) / derivFunc(b,x)      
            y= x
            x = x - h
            z= x
            E=abs(((z-y)/z)*100)

            fa.append(i)
            fd.append(E)
    
            print(''+str(i)+'    |\t'+str('%6f'%x)+'\t|\t'+str(func(a,x))+'\t|\t'+str(derivFunc(b,x))+'\t|\t'+str('%6f'%(E))+'\t|')
            
            # Plot between Func(x) vs X
            ax.scatter(x,func(a,x))
            
            # set a title and labels
            ax.set_title('F(x) vs x')
            ax.set_xlabel('x')
            ax.set_ylabel('F(x)')
  
            i+=1

            if E<e:
                break
            else:
                continue
    return fa,fd


def Secant(x,x1,a,c,e):
    fa=[]
    fd=[]
    i=1
    E=0
    print('I. No '+'\t|\t'+'X0'+'\t|\t'+'X1'+'\t\t|\t'+'F(x0)''\t\t|\t'+'F(x1)'+'\t\t|\t'+'Ea'+'\t\t|')
    
    while i<=c:
        
        
        if i==1:
            print(''+str(i)+'\t|\t'+str('%6f'%x)+'|\t'+str('%6f'%x1)+'\t|'+str(func(a,x))+'\t|'+str(func(a,x1))+'\t|\t'+str('%6f'%(E))+'\t|')
            
            # Plot between Func(x) vs X
            ax.scatter(x,func(a,x))
            
            # set a title and labels
            ax.set_title('F(x) vs x')
            ax.set_xlabel('x')
            ax.set_ylabel('F(x)')
            fa.append(i)
            fd.append(E)
            i+=1

        else: 
            
            h=(x-x1)/(func(a,x)-func(a,x1))
            x=x1
            x1=x1-(func(a,x1)*h)
            E=abs((x1-x)/x1*100)
            print(''+str(i)+'\t|\t'+str('%6f'%x)+'|\t'+str('%6f'%x1)+'\t|'+str(func(a,x))+'\t|'+str(func(a,x1))+'\t|\t'+str('%6f'%(E))+'\t|')
            
            # plot between Func(x) vs X
            ax.scatter(x,func(a,x))
            
            # set a title and labels
            ax.set_title('F(x) vs x')
            ax.set_xlabel('x')
            ax.set_ylabel('F(x)')
            fa.append(i)
            fd.append(E)
            i+=1
            
            if E<e:
                break
            else:
                continue
            
    return fa,fd          

# INPUTS
p=int(input('Choose 1 for Newton Raphson Method and 2 for Secant Method ='))
a= input("Write the value of Function = ")
a=sy.sympify(a)
b=sy.diff(a)
print("so our derivative is = ",b)
c= input("Write the no of iteration required = ")   
c=int(c)
e= input(" Write the max relative approximate error (%) = ")
e=float(e)
x = input(" Write value of intial guess x0 = ")
x=float(x)
if p==1:
    
    u,v=newtonRaphson(x,a,b,c,e)
elif p==2:
    
    x1=input(" Write value of intial guess x1 = ")
    x1=float(x1)
    u,v=Secant(x,x1,a,c,e)
else:
    print('Wrong Input')

# CREATED BY SWAPIL VERMA 
# FOR ANY SUGGESTION OR COLLABORATION FEEL FREE TO CONTACT ME ON LINKEDIN
# https://www.linkedin.com/in/swapil-verma-6b787a165/
```
